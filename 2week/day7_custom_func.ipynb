```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("titanic")
print(df.shape)
print(df.info())
```

```python
df.describe(include='all')
```

```python
df.head(10)
```

```python
# apply()
# 1. class 만들기
df.drop('class', axis=1, inplace=True)
df.head(10)
```

```python
def make_class(pclass):
    if pclass == 1:
        return 'First'
    elif pclass == 2:
        return 'Second'
    else:
        return 'Third'
```

```python
df['class'] = df['pclass'].apply(make_class)
df.head(10)
```

```python
# 2.티겟값 할인가인지 확인하기
# 각 class별 평균가 이하면 할인가 라고 가정

mean_fare_by_class = df.groupby('pclass')['fare'].mean().values
print(mean_fare_by_class)

def discounted(df):
    if df['pclass'] == 1:
        if df['fare'] < mean_fare_by_class[0]:
            return True
        else:
            return False
        
    elif df['pclass'] == 2:
        if df['fare'] < mean_fare_by_class[1]:
            return True
        else:
            return False
        
    elif df['pclass'] == 3:
        if df['fare'] < mean_fare_by_class[2]:
            return True
        else:
            return False
```

```python
df['Discount'] = df.apply(discounted, axis=1)
df.head(10)
```

```python
# map()
df.drop('class', axis=1, inplace=True)
df.head(5)
```

```python
class_mapping = {1:'First', 2:'Second', 3:'Third'}

df['class'] = df['pclass'].map(class_mapping)
df.head(5)
```

```python
# 특정값만 매핑도 가능( 없는값 null처리)
embarked_mapping = {'S':True}

df['embarked_mapping'] = df['embarked'].map(embarked_mapping)
df.head(5)
```

```python
df.drop('embarked_mapping', axis=1, inplace=True)
df.head(5)
```

```python
# lambda
df['vat'] = df['fare'].apply(lambda x: x*0.1)
df.head()
```

```python
df.drop('vat', axis=1, inplace=True)
```

```python
df['vat'] = df.apply(lambda x: x['fare']*0.1 if x['class'] == 'First' else x['fare']*0.15, axis=1)
df.head()
```

```python
# 추가 실습
df= sns.load_dataset("titanic")

# 1. 나이별 그룹화
def age_group(age):
    if age <= 18:
        return 'Child'
    elif age <= 35:
        return 'Young Adult'
    elif age <=  60:
        return 'Adult'
    else:
        return 'Senior'
    
df['age_group'] = df['age'].apply(age_group)
df.head()
```

```python
# 2. 생존 확률 평가
df['survival_probability'] = df.apply(lambda x: 'High' if (x['age_group'] == 'Child') or (x['age_group'] == 'Senior')
                                      or (x['pclass'] == 1) else 'Low', axis=1)

df[['age_group', 'pclass', 'survival_probability']].head()
```

```python
# 3. 가족 수 계산
df['family_size'] = df.apply(lambda x: x['sibsp'] + x['parch'], axis=1)
df[['sibsp', 'parch', 'family_size']].head(10)
```
