```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("titanic")
df.shape
```

```python
df.info
```

```python
df.head()
```

```python
# groupby
# 1. 성별에 따른 생존률 평균
df.groupby('sex')['survived'].mean()
```

```python
# 2.클래스별 요금과 생존률 평균
df.groupby('class')[['fare', 'survived']].mean()
```

```python
# 3.성별 + 클래스별 생존률
df.groupby(['class', 'sex'])['survived'].mean()
```

```python
# 4. 티겟 값의 합과 중앙값
print(df.groupby('class')['fare'].sum())
print(df.groupby('class')['fare'].median())
```

```python
# 5.groupby에 discribe 사용해보기
df.groupby(['class', 'embark_town'])['survived'].describe()
```

```python
# groupby 고급
# 1. agg
df.groupby('sex')['survived'].agg(['mean', 'std', 'count'])
```

```python
# 2. 필터링
def filtering(df):
    return df['age'].mean() > 30

df.groupby('class').filter(filtering)['class'].unique()
```

```python
# 3. transform
df['mean_fare_by_pclass'] = df.groupby('pclass')['fare'].transform('mean')
df[['pclass', 'fare', 'mean_fare_by_pclass']].head(10)
```

```python
# pivot_table
# 1. 성별 + 클래스별 생존율
pt = pd.pivot_table(df, values='survived', index='class', columns='sex', aggfunc='mean')
pt
```

```python
# 2. 생존자들의 클래스별 평균 요금
pt = pd.pivot_table(df, values='fare', index='survived', columns='class', aggfunc='mean')
pt
```

```python
# 3. 결측치 채우기
pt_filled = pd.pivot_table(df, values='survived', index='class', columns='sex', aggfunc='mean', fill_value=0)
pt_filled
```

```python
# 추가 실습
# 1. 탑승 마을별 생존률 평균
df.groupby('embark_town')['survived'].mean()
```

```python
# 2. 나잇대 구간화 해보기
def get_age_range(age):
    return age // 10 * 10

df['age_range'] = df['age'].transform(get_age_range)
df[['age', 'age_range']].head(10)
```

```python
df.groupby('age_range')['survived'].mean()
```

```python
# 3. 다중 피벗테이블 만들기
pt = pd.pivot_table(df, values='survived', index=['class', 'sex', 'embark_town'], aggfunc='sum')
pt
```
