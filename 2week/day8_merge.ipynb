```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('diamonds')
print(df.shape)
print(df.info())
```

```python
df.head()
```

```python
df.describe()
```

```python
# merge
cut_avg_price = df.groupby('cut')['price'].mean().reset_index()
cut_avg_price.columns=['cut', 'avg_price']
cut_avg_price.head()
```

```python
df_merge = pd.merge(df, cut_avg_price, on='cut', how='left')
df_merge.head()
```

```python
cut_color_avg_depth = df.groupby(['cut', 'color'])['depth'].mean().reset_index()
cut_color_avg_depth.columns=['cut', 'color', 'avg_depth']
cut_color_avg_depth.head()
```

```python
df_merge2 = pd.merge(df_merge, cut_color_avg_depth, on=['cut', 'color'], how='left')
df_merge2.head()
```

```python
# 추가 실습
# 1. 평균 추가 병합
clarity_avg_carat = df.groupby('clarity')['carat'].mean().reset_index()
clarity_avg_carat.columns = ['clarity', 'carat_mean']
clarity_avg_carat.head()
```

```python
df_merge3 = pd.merge(df_merge2, clarity_avg_carat, on='clarity', how='left')
df_merge3
```
